```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv("data/train.csv")
```

```python
df_train.head()
```

```python
df_train.columns
```

```python
cols_to_keep = ["ExterQual", "LotArea", "YrSold", "SalePrice"]
```

```python
df = df_train[cols_to_keep]
```

```python
# Calculate the number of missing values in each column
missing_values_summary = df[cols_to_keep].isna().sum()

# Display the summary
print(missing_values_summary)
```

```python
# Plot a histogram of the 'SalePrice' column
plt.hist(df['SalePrice'], bins=10, color='blue', edgecolor='black')

# Add labels and title
plt.xlabel('Sale Price')
plt.ylabel('Frequency')
plt.title('Histogram of Sale Price')

# Show the plot
plt.show()
```

```python
import seaborn as sns
```

```python
# Create a FacetGrid plot
g = sns.FacetGrid(df, col="YrSold", col_wrap=3, height=4)
g.map(sns.histplot, "SalePrice", bins=10, color='blue', edgecolor='black')

# Add labels and title
g.set_axis_labels('Sale Price', 'Frequency')
g.set_titles('Year Sold {col_name}')

# Show the plot
plt.show()
```

```python
# Create a scatterplot
sns.scatterplot(x='LotArea', y='SalePrice', data=df, color='blue')

# Add labels and title
plt.xlabel('Lot Area')
plt.ylabel('Sale Price')
plt.title('Scatterplot of Sale Price vs Lot Area')

# Make the y-axis logarithmic
plt.xscale('log')

# Show the plot
plt.show()
```
